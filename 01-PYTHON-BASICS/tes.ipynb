```python
def even_or_odd(num):
    """"Find a number is even or odd"""
    if (num%2  == 0):
        isNum = "even"
        print("Number is even")
    else:
        isNum="odd"
        print("Number is odd")
    return print(f"{num} is {isNum}")

```

```python
#csv
import csv
with open("example.csv", mode="w", newline='') as file:
    writer=csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(["santosh", 32])

with open('example.csv', mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## Class, object, methods, attributes
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def bark(self):
        print(f"{self.name} says boof , bhown bhown")

dog1 = Dog("Moti", 10)
dog2 = Dog("Tiger", 7)
dog1.bark()
dog2.bark()

        
```
